```python
import re

txt = "Abraham Lincoln, Andrew P Garfield, Connor Milliken, Jordan Alexander Williams, Madonna, programming is cool"
pattern = re.compile('([A-Z][a-zA-Z]+) ([A-Z][a-z]+)')
pattern2 = re.compile('([A-Za-z][A-Za-zA-Za-z]+) ([A-Za-z]+) ([A-Z][A-Za-z]+)')
found_names = pattern.findall(txt)
found = pattern2.findall(txt)
fun = str([found_names[0], found_names[1], found])
print(fun)

    

```

```python
import re

f = open("names.txt")
data = f.read()
print(data)
    
```

```python
example = re.compile('([A-Z][a-zA-Z]+[\W]+) ([A-Z][a-z]+)')
found = example.findall(data)
twitter_username_re = re.compile('@([A-Za-z]+[A-Za-z0-9-_]+)')
twitter_found = twitter_username_re.findall(data)
print(twitter_found)
print(found)
```
